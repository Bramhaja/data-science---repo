```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Simple dataset: Hours studied vs. Marks scored
data = {
    'Hours': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'Scores': [30, 35, 50, 55, 60, 65, 70, 80, 90]
}
df = pd.DataFrame(data)
print(df)

plt.scatter(df['Hours'], df['Scores'], color='blue')
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.title("Study Hours vs Marks")
plt.show()

X = df[['Hours']]  # independent variable (2D array)
y = df['Scores']   # dependent variable

model = LinearRegression()
model.fit(X, y)

# Predict for 5.5 hours
hours = np.array([[5.5]])
predicted_score = model.predict(hours)
print(f"Predicted Score for 5.5 study hours = {predicted_score[0]:.2f}")

plt.scatter(df['Hours'], df['Scores'], color='blue')
plt.plot(df['Hours'], model.predict(X), color='red')
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.title("Linear Regression - Study Hours vs Marks")
plt.show()
```
