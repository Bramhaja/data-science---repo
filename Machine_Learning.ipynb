```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df = pd.read_csv('student_scores.csv')

X = df.iloc[:, :-1].values
y = df.iloc[:, 1].values

X_train, X_test, y_train, y_test = train_test_split(X, y,
test_size=0.2, random_state=0)

regressor = LinearRegression()

regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

d = {'Actual': y_test, 'Predicted': y_pred}

compare_df = pd.DataFrame(d)
print(compare_df)
```

```python
x_pred = regressor.predict(y_test)

d = {'Actual': X_test, 'Predicted': x_pred}

compare_df = pd.DataFrame(d)
print(compare_df)
```

```python
X_input_reshaped = y_test.reshape(-1, 1)

# 2. Pass the reshaped 2D data to the predict method
y_pred = regressor.predict(X_input_reshaped)

# The rest of your code works fine now:
d = {'Actual': y_test, 'Predicted': y_pred}
compare_df = pd.DataFrame(d)
print(compare_df)
```
